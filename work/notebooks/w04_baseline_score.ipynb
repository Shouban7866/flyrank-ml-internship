## 2. Build the ranked queue (writes the CSV)

*Code the score, rank everything, write work/outputs/baseline_action_score.csv.*

```python
import pandas as pd
import os

df = pd.read_csv("/content/sample_data/california_housing_test.csv")

# Baseline Score
df["baseline_score"] = df["median_income"]

# Rank
df = df.sort_values("baseline_score", ascending=False)

# Reason Codes
def reason(x):
    if x >= 6:
        return "RC1"
    elif x >= 3:
        return "RC2"
    else:
        return "RC3"

df["reason_code"] = df["baseline_score"].apply(reason)

os.makedirs("work/outputs", exist_ok=True)

df.to_csv("work/outputs/baseline_action_score.csv", index=False)

print(df[["median_income","baseline_score","reason_code"]].head(20))
```
