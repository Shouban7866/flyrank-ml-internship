## 2. Build the ranked queue (writes the CSV)

*Code the score, rank everything, write work/outputs/baseline_action_score.csv.*

```python
import pandas as pd
import os

df = pd.read_csv("/content/content_refresh_anonymized.csv")

# Fill missing values
df["impressions_90d"] = df["impressions_90d"].fillna(0)
df["ctr"] = df["ctr"].fillna(0)

# Simple baseline score
df["baseline_score"] = df["impressions_90d"] + (df["ctr"] * 100)

# Rank
df = df.sort_values("baseline_score", ascending=False)

# Reason codes
def reason(row):
    if row["impressions_90d"] >= 10000 and row["ctr"] >= 1:
        return "RC1"
    elif row["impressions_90d"] >= 1000:
        return "RC2"
    else:
        return "RC3"

df["reason_code"] = df.apply(reason, axis=1)

os.makedirs("work/outputs", exist_ok=True)

df.to_csv("work/outputs/baseline_action_score.csv", index=False)

print(df[["content_id", "baseline_score", "reason_code"]].head(20))
```
