## 5. Exports for the paper

*Write the queue (and any figures you want to reuse) to work/outputs/ — your paper builds on these files.*

The ranked content action queue is exported to work/outputs/content_action_queue.csv.

The exported file contains the recommended action and reason code for each content item. The recommendations are based on observed SEO and engagement signals and are intended for decision support. The exported file can be reused in the final report.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
```

```python
import pandas as pd
import os

# Load dataset
df = pd.read_csv("/content/content_refresh_anonymized.csv")

# Create action queue
action_queue = df[
    [
        "content_id",
        "impressions_90d",
        "ctr",
        "engagement_rate",
        "content_age_days"
    ]
].copy()

# Generate recommended actions
def recommend_action(row):
    if row["engagement_rate"] < 2:
        return "Improve Low Engagement"
    elif row["content_age_days"] > 365:
        return "Update Old Content"
    elif row["ctr"] >= 1 and row["engagement_rate"] >= 2:
        return "Promote High Performing Content"
    else:
        return "Review Declining Content"

action_queue["recommended_action"] = action_queue.apply(
    recommend_action,
    axis=1
)

# Assign reason codes
reason_map = {
    "Improve Low Engagement": "LOW_ENGAGEMENT",
    "Update Old Content": "CONTENT_STALE",
    "Promote High Performing Content": "HIGH_VALUE",
    "Review Declining Content": "PERFORMANCE_DROP"
}

action_queue["reason_code"] = action_queue["recommended_action"].map(reason_map)

# Export results
os.makedirs("work/outputs", exist_ok=True)

output_path = "work/outputs/content_action_queue.csv"

action_queue.to_csv(output_path, index=False)

print(action_queue.head())
print("\nExport saved:", output_path)
```
