## Ranked Actions + Reason Codes

The action queue prioritizes content improvement tasks based on observed performance signals.

| Rank | Action | Reason Code | Explanation |
|---|---|---|---|
| 1 | Improve low engagement content | LOW_ENGAGEMENT | Content receives less interaction and needs optimization |
| 2 | Update old content | CONTENT_STALE | Older content may contain outdated information |
| 3 | Promote high performing content | HIGH_VALUE | Content shows positive engagement signals |
| 4 | Review declining content | PERFORMANCE_DROP | Content performance is decreasing |

These recommendations are decision-support suggestions and require human review before action.
